# 1. Introduction

RQ: What was the influence of thinness among children and adolescents aged 1 to 19 on the overall BMI in developed countries for the year 2012 and was this influence different for developing countries?




# 2. Data preparation

```python
import pandas as pd
dataset = pd.read_csv('https://raw.githubusercontent.com/NHameleers/dtz2025-datasets/master/CountryHealthFactors.csv')
dataset.head()
```

```python
dataset = dataset.rename(columns=str.strip)
dataset.BMI.describe()
```

# 3. Explore and clean the Data

```python
dataset.isnull().sum()
```

```python
dataset = dataset.rename(columns=str.strip)
dataset2 = dataset.filter(['Country', 'Year', 'Status', 'BMI', 'thinness  1-19 years', 'thinness 5-9 years'], axis = 1)
dataset2.count()
```

```python
dataset2 = dataset2.dropna(thresh=6, subset=['Country', 'Year', 'Status', 'BMI', 'thinness  1-19 years', 'thinness 5-9 years'])

mean_developing = dataset2[dataset2['Status'] == 'Developing']['BMI'].mean()
mean_developed = dataset2[dataset2['Status'] == 'Developed']['BMI'].mean()

dataset2.loc[dataset2['Status'] == 'Developing', 'mean_Developing'] = mean_developing
dataset2.loc[dataset2['Status'] == 'Developed', 'mean_Developed'] = mean_developed
dataset2.tail(50)

dataset2 = dataset2[(dataset2['BMI'] >= 10) & (dataset2['BMI'] <= 40)]
dataset2
```

```python
dataset2 = dataset.loc[dataset.Year == 2012, :]
dataset2.head(100)
```

# 4. Describe and Visualize

```python
import altair as alt
import ipywidgets as widgets
from ipywidgets import interact

dataset2 = pd.DataFrame(dataset2)

def select_data(Status, dataset2):
  '''Takes Status and dataset2 as input. Returns the rows in
  dataset belonging to country with Status'''

  return dataset2.loc[dataset2['Status'] == Status, ]

def visualize_data(Status, dataset2):
  data_to_viz = select_data(Status, dataset2)
  correlation_coefficient = dataset2['BMI'].corr(dataset2['thinness  1-19 years'])
  plot = alt.Chart(data_to_viz).mark_point().encode(
      x='BMI:Q',
      y='thinness  1-19 years:Q',
  )
  regression_line = plot.transform_regression('BMI', 'thinness  1-19 years').mark_line().encode(
        tooltip=[alt.Tooltip('correlation_coefficient:Q', title='Correlation Coefficient')]
    )
  chart = (plot + regression_line)
  display(chart)

interact(visualize_data,
         Status=['Developing', 'Developed'],
         dataset2=widgets.fixed(dataset2))


```
